```python
dat = pd.read_csv('45_vars.csv')
dat = dat.drop(columns="Unnamed: 0")
```

### Z-scale each variable

```python
# z scale each variable
dat_scaled = StandardScaler().fit_transform(dat)
dat = pd.DataFrame(dat_scaled, columns = dat.columns)
```

```python
dat.head(10).transpose()
```

### PCA

```python
pca = PCA(0.9)
principalComponents_dat = pca.fit_transform(dat)
```

```python
principalComponents_dat
```

```python
pca_dat = pd.DataFrame(data = principalComponents_dat, columns = ['principal component 1', 'principal component 2', 'principal component 3', 'principal component 4', 'principal component 5'])
```

```python
pca_dat
```

### Z-scale again

```python
# z scale again
dat_scaled_2 = StandardScaler().fit_transform(pca_dat)
pca_dat_scaled = pd.DataFrame(dat_scaled_2, columns = ['z-score 1', 'z-score 2', 'z-score 3', 'z-score 4', 'z-score 5'])
```

```python
pca_dat_scaled.shape
```

```python
pca_dat_scaled
```

### Score 1

```python
# create score 1
temp = np.square(np.absolute(pca_dat_scaled))
```

```python
temp
```

```python
s1 = (temp.sum(axis = 1))**(1/2)
s1
```

### Score 2

```python
# create score 2
input_dim = pca_dat_scaled.shape[1]  # input_dim = 5
hidden_dim = 3
input_layer = Input(shape=(input_dim, ))
```

```python
hidden = Dense(hidden_dim)(input_layer)
output = Dense(input_dim)(hidden)
```

```python
autoencoder = Model(inputs = input_layer, outputs = output)
autoencoder.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = ['accuracy'])
autoencoder.fit(pca_dat_scaled, pca_dat_scaled, epochs = 1)
```

```python
pred = autoencoder.predict(pca_dat_scaled)
mean_squared_error(pred, pca_dat_scaled)
```

```python
pred
```

```python
temp2 = pd.DataFrame(pred - pca_dat_scaled.values)
```

```python
temp2 = np.square(np.absolute(temp2))
```

```python
temp2
```

```python
s2 = (temp2.sum(axis = 1))**(1/2)
s2
```

```python
# Append the two scores to the original dataframe
dat['s1'] = s1 
dat['s2'] = s2 
```

```python
dat
```

### Create Final Score

```python
# Extreme Quantile Binning
dat['s1_rank'] = dat['s1'].rank(ascending = False)
dat['s2_rank'] = dat['s2'].rank(ascending = False)
```

```python
dat['final_score'] = (dat['s1_rank'] + dat['s2_rank'])/2
```
