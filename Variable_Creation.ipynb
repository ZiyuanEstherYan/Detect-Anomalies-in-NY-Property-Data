```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mydata = pd.read_csv('NY property data.csv')
```

```python
clean_data = mydata
```

```python
# rename ZIP to ZIP5
clean_data.rename(columns={'ZIP':'ZIP5'}, inplace=True)
```

```python
# create new variable ZIP3
clean_data['ZIP3'] = mydata['ZIP5'].astype(str).str[0:3].astype(float)
```

```python
clean_data.head(10).transpose()
```

```python
clean_data.tail(10).transpose()
```

```python
# fill in missing values by B
boro_mean = clean_data.groupby('B').mean()
```

```python
boro_mean.head(10).transpose()
```

```python
# fill in missing values of FULLVAL
if clean_data['FULLVAL'] == 0:
    if clean_data['B'] == 1:
        clean_data['FULLVAL'] = boro_mean['FULLVAL'][1]
    elif clean_data['B'] == 2:
        clean_data['FULLVAL'] = boro_mean['FULLVAL'][2]
    elif clean_data['B'] == 3:
        clean_data['FULLVAL'] = boro_mean['FULLVAL'][3]
    elif clean_data['B'] == 4:
        clean_data['FULLVAL'] = boro_mean['FULLVAL'][4]
    elif clean_data['B'] == 5:
        clean_data['FULLVAL'] = boro_mean['FULLVAL'][5]
```

```python
'NaN' in clean_data['FULLVAL']
```

```python
V1 = clean_data['FULLVAL']
V2 = clean_data['AVLAND']
V3 = clean_data['AVTOT']
```

```python
S1 = np.multiply(clean_data['LTFRONT'], clean_data['LTDEPTH'])
S2 = np.multiply(clean_data['BLDFRONT'], clean_data['BLDDEPTH'])
S3 = np.multiply(S2, clean_data['STORIES'])
```

```python
r1 = V1/S1
r2 = V1/S2
r3 = V1/S3
r4 = V2/S1
r5 = V2/S2
r6 = V2/S3
r7 = V3/S1
r8 = V3/S2
r9 = V3/S3
```
